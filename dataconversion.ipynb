```python
import cv2
vidcap = cv2.VideoCapture('./Data/inazuma_amakumo_peak.mp4')
success,image = vidcap.read()
count = 0
while success:
  cv2.imwrite("./Data/amakumo_peak/frame%d.jpg" % count, image)     # save frame as JPEG file      
  success,image = vidcap.read()
  print('Read a new frame: ', success)
  count += 1
```
